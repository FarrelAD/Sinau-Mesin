# Preparation

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

# Data Overview

```python
df = pd.read_csv('../../datasets/wisconsin-breast-cancer.csv')
df.head()
```

```python
df.describe()
```

# Goal

To diagnose **malignant (M)** and **benign (B)** types of cancer.

# 1. Separate between usable and unusable variables.

Based on 32 columns, it needs to remove column that have pattern `id` and `Unnamed: 32`.

```python
df_clean = df.drop(['id', 'Unnamed: 32'], axis=1, errors='ignore')
df_clean.head()
```

```python
df.describe()
```

Usable features

```python
useful_features = df_clean.select_dtypes(include=['float64', 'int64']).columns
useful_features
```

Unusable features

```python
print(['id', 'Unnamed: 32'])
```

# 2. Perform the encoding process in the "diagnosis" column.

Previous diagnosis column overview

```python
df['diagnosis'].head()
```

```python
df['diagnosis'].describe()
```

Current diagnosis column overview

```python
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

df.head()
```

```python
df['diagnosis'].describe()
```

# 3. Perform the standardization process on all columns that have numeric values.

```python
numeric_cols = df.select_dtypes(include=["number"]).columns

scaler = StandardScaler()

df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

df.head()
```
